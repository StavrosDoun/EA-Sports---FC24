```python
df_after_drop['value'] = df_after_drop['value'].str.replace('$', '').replace('.', '')
df_after_drop['value']= df_after_drop['value'].str.replace('.','')
df_after_drop
```

```python
df_after_drop['value'] = df_after_drop['value'].apply(pd.to_numeric)
```

```python
df_after_drop['value'].info()
```

```python
df_after_drop.shape
```

```python
df_after_drop_2 = df_after_drop.drop(['player', 'country', 'club'], axis = 1)
```

```python
df_after_drop_2.shape
```

```python
X = df_after_drop_2.drop(['value'], axis = 1)
y = df_after_drop_2['value']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state = 0)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
r2_score(y_test, y_pred)
```

```python
# run a linear to the dataset
model = LinearRegression()
model.fit(X_train, y_train)
y_pred_lin = model.predict(X_test)
print('Mean Squared Error :', mean_squared_error(y_test, y_pred_lin))
print('R2 score : ', r2_score(y_test, y_pred_lin))
```
